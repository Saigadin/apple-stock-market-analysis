```python
data.rename(columns = {
    'AAPL.Open' : 'Open',
    'AAPL.High' : 'High',
    'AAPL.Low' : 'Low',
    'AAPL.Volume' : 'Volume',
    'AAPL.Adjusted' : 'Adjusted',
    'AAPL.Close' : 'Close'
}, inplace = True)
```

```python
data['Date'] = pd.to_datetime (data['Date'])
```

```python
data.isnull().sum()
```

```python
data['Daily_return'] = data['Close'].pct_change()
```

```python
data['MA30'] = data['Close'].rolling(window=30).mean()
```

```python
data['Price_Change'] = data['Close'] - data['Open']
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.dropna()
```

```python
data = data[['Date','Open','High','Low','Close','Volume','Daily_return','MA30','Price_Change']]
```

```python
data.to_csv("cleaned_AAPL_stock.csv", index = False)
```

```python
import matplotlib.pyplot as plt
```

STOCK PRICE TREND

```python
plt.figure(figsize=(12,6))

data['Date'] = pd.to_datetime (data['Date'])

plt.plot(data['Date'], data['Close'])

plt.title("Apple Stock Closing Price Trend")

plt.xlabel("Date")

plt.ylabel("Closing price")

plt.show()
```

MOVING AVERAGE TREND

```python
plt.figure(figsize=(12,6))

plt.plot(data['Date'],data['Close'], label = 'Closing Price')

plt.plot(data['Date'],data['MA30'], label = '30 Day Moving Average')

plt.title("Stock Price vs Moving Average")

plt.xlabel("Date")

plt.ylabel("Price")

plt.legend()

plt.show()
```

TRADING VOLUME ANALYSIS

```python
plt.figure(figsize=(12,6))

plt.bar(data['Date'],data['Volume'])

plt.title("Daily Trading Vaolume")

plt.xlabel("Date")

plt.ylabel("Volume")

plt.show()
```

PRICE VS VOLUME

```python
plt.figure(figsize=(12,6))

plt.scatter(data['Close'],data['Volume'])

plt.title("Stock price vs Trading volume")

plt.xlabel("Closing Price")

plt.ylabel('Volume')

plt.show()
```

DAILY RETURN DISTRIBUTION

```python
plt.figure(figsize=(12,6))

plt.hist(data['Daily_return'],bins =10)

plt.title("Daily Return Distribution")

plt.xlabel("Daily Return")

plt.ylabel("Frequency")

plt.show()
```
